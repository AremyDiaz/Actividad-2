```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga del archivo.csv
data = pd.read_csv("airbnb - Brazil.csv") #, index_col=0
data
```

Reemplazo de nulos por columna

```python
#Copia del dataset original
data2 = data.copy()
data2
```

```python
#Cuántos nulos hay por columna
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir los nulos de name, host_name, neighbourhood_group y license por el string "Desconocido"
data2["name"]=data2["name"].fillna("Unknown")
data2["host_name"]=data2["host_name"].fillna("Unknown")
data2["neighbourhood_group"]=data2["neighbourhood_group"].fillna("Unknown")
data2["license"]=data2["license"].fillna("Unknown")
```

```python
#Sustituir los nulos de last_review con el string "No_reviews"
data2["last_review"]=data2["last_review"].fillna("No_reviews")
```

```python
#Sustituir los nulos de number_of_reviews con el valor 0 porque son casos en los que NO HAY REVIEWS
data2["number_of_reviews"]=data2["number_of_reviews"].fillna(0)
data2["reviews_per_month"]=data2["reviews_per_month"].fillna(0)
```

```python
#Comprobar que ya no haya datos faltantes
#Cuántos nulos hay por columna
valores_nulos=data2.isnull().sum().sum()
valores_nulos
```

Sustitución de atípicos por distribución normal

```python
#Copia del dataset sin nulos
data3 = data2.copy()
data3
```

```python
#Valores extremos: límite superior y límite inferior.
#Variable: minimum_nights.
x1=data3["minimum_nights"]
LS1= x1.mean() + 3*x1.std()
LI1= x1.mean() - 3*x1.std()
print("Limite superior permitido", LS1)
print("Limite inferior permitido", LI1)
```

```python
#Outliers de minimum_nights.
outliers_minimum_nights = data3[(x1>LS1)|(x1<LI1)]
len(outliers_minimum_nights)
```

```python
#Primera sustitución: minimum_nights.
data3=data2[(x1<=LS1)&(x1>=LI1)] #Obtenemos datos sin outliers de price del DataFrame
data3["minimum_nights"]=data3["minimum_nights"].fillna(round(data2["minimum_nights"].mode(),0))
```

Generar .csv con datos limpios

```python
#Convertir DataFrame limpio a CSV
clean_airbnb=data3.copy()
clean_airbnb.to_csv("clean_airbnb - Brazil.csv")
```
